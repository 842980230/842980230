**# Correlação entre idade e probabilidade de ser devedor**


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot para visualizar a relação entre idade e ser devedor
plt.figure(figsize=(8, 6))
sns.boxplot(x='Cliente_Devedor?', y='Idade', data=base)
plt.title('Relação entre Idade e Probabilidade de Ser Devedor')
plt.xlabel('Cliente Devedor?')
plt.ylabel('Idade')
plt.show()

# Convert 'Cliente_Devedor?' to numeric type (assuming it's categorical)
base['Cliente_Devedor_Num'] = base['Cliente_Devedor?'].map({'Yes': 1, 'No': 0}) # Replace 'Yes' and 'No' with actual values if different

# Calcular a correlação entre as variáveis
correlacao = base['Idade'].corr(base['Cliente_Devedor_Num'])
# Uso da nova coluna numerica
print("Correlação entre Idade e Cliente Devedor:", correlacao)
```

**Tabela de contingência entre profissão e compra do título**

```python
import pandas as pd

# Criar uma tabela de contingência
tabela_contingencia = pd.crosstab(base['Profissão'], base['Cliente_Comprou_o_Titulo?'])
print(tabela_contingencia)

# Calcular o chi-quadrado para testar a independência entre as variáveis
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(tabela_contingencia)
print("Valor de p:", p)
```

**Histograma do saldo da conta corrente, segmentado por estado civil**

```python
import matplotlib.pyplot as plt

# Criar um histograma para cada estado civil
plt.figure(figsize=(10, 6))
for estado_civil in base['Estado_Civil'].unique():
    subset = base[base['Estado_Civil'] == estado_civil]
    plt.hist(subset['Saldo_Conta_Corrente'], bins=10, alpha=0.5, label=estado_civil)

plt.title('Distribuição do Saldo da Conta Corrente por Estado Civil')
plt.xlabel('Saldo da Conta Corrente')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```
